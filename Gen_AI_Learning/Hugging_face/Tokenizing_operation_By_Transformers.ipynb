| `DistilBertTokenizer`                  | A specific tokenizer for the DistilBERT model, used to convert text into a format that the DistilBERT model can understand.             |
| `DistilBertForSequenceClassification`| A specific model class for the DistilBERT model fine-tuned or ready for sequence classification tasks.                                   |

# Lets try to see how tokenzer is working

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('google-bert/bert-base-uncased')


#example text
text = "barbie girl ummah"
tokens = tokenizer.tokenize(text)
print("tokens:",tokens)
```

# Convert Tokens To input id's

```python
#here we take the above tokens to conver

input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("input_id's :",input_ids)
```

# Encoding

```python
 #encoding = converting the text tokenization + converting the input_id's
encoded_input = tokenizer(text)
print("Encoded_input:",encoded_input)

```

```python
#decode The text

decoded_output = tokenizer.decode(input_ids)
print("Decode_output:",decoded_output)
```
