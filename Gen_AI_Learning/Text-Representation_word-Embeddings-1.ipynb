# **TF-IDF (Term frequency- Inverse document frequency)**

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

```

```python
arr = tfidf.fit_transform(df['text']).toarray()
```

```python
arr
```

```python
print(tfidf.idf_)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Example corpus of 100 docs (we'll just simulate with repeats)
docs = ["machine learning is fun"] * 2 + ["other text"] * 98

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

# Get IDF value of "machine"
idf_value = dict(zip(vectorizer.get_feature_names_out(), vectorizer.idf_))["machine"]

print("IDF(machine):", idf_value)
```
